```r
conditions <- read.table(
    '../../../datasets/01-simulation01/conditions.tsv', 
    header=TRUE
)
```

```r
set.seed(8923767)

for (row_index in 1:nrow(conditions)) {
    cond <- conditions[row_index,]
    for (replicate in 1:10) {
        # Yule process with lambda = 1
        phy <- ladderize(rphylo(n = cond$ntips, birth = 1, death = 0))
        phy <- makeNodeLabel(phy, prefix = "i")
        
        file <- sprintf("./rphylo/%s-%02d.tree", cond$conditionID, replicate)
        write.tree(phy = phy, file = file)
    }
}
```
